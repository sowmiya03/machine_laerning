```python
def generate_keywords(doc):
  r = Rake()
  r.extract_keywords_from_text(doc)
  keys=r.get_ranked_phrases()
  l=[]
  for i in keys:
    li=i.split()
    for j in li:
      if j.find('ing') == -1 and j[-1]!='d' and j[-2]!='e':
        l.append(j)
  return l
```
